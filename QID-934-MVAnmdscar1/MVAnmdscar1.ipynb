```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[3, 2, 1, 10], [2, 7, 3, 4]]).T
brands = ["Mercedes", "Jaguar", "Ferrari", "VW"]

plt.subplots(figsize=(7, 7))
plt.scatter(x[:, 0], x[:, 1], c="b", s=60)
for i in range(len(x)):
    plt.text(x[i, 0] + 0.05, x[i, 1] + 0.1, brands[i], fontsize=14)

plt.xlim(0, 12)
plt.ylim(0, 8)
plt.title("Initial Configuration", fontweight="heavy", fontsize=14)
plt.tight_layout()
plt.savefig("MVAnmdscar1.png", transparent=True, dpi=144)
plt.show()
```
