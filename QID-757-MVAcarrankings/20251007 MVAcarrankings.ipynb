```python
from sklearn.isotonic import IsotonicRegression
import matplotlib.pyplot as plt
```

```python
# the reported preference orderings
x = list(range(1, 7))
# the estimated preference orderings according to the additive model (16.1) and 
# the metric solution (Table 16.6) in MVA
y = [0.84, 2.84, 3.16, 3.34, 5.66, 5.16]
```

```python
gp = IsotonicRegression()
y_gp = gp.fit_transform(x, y)
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
ax.plot(x, y_gp, c = "k")
ax.scatter(x, y, c = "r")
for i in range(0, len(y)):
    ax.text(x[i]-0.05, y_gp[i]+0.1, "car" + str(i+1), fontsize = 12)
plt.xlabel("Revealed rankings", fontsize = 12, fontweight='heavy')
plt.ylabel("Estimated rankings", fontsize = 12, fontweight='heavy')
plt.title("Car rankings", fontsize = 14, fontweight='heavy')
plt.tight_layout()
plt.savefig("MVAcarrankings.png", dpi=144, transparent=True)
plt.show()
```
