```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[3, 2], [2, 7], [1, 3], [0.37, 4.36], [10, 4]])
brands = ["Mercedes", "Jaguar", "Ferrari Init", "Ferrari New", "VW"]
```

```python
plt.figure(figsize=(7, 7))
plt.scatter(x[:, 0], x[:, 1], c="r")
for i in range(len(x)):
    plt.text(x[i, 0] + 0.05, x[i, 1] + 0.1, brands[i], fontsize=14)
plt.plot(x[2:4, 0], x[2:4, 1], c="r")

plt.xlim(0, 12)
plt.ylim(0, 8)
plt.title("First iteration for Ferrari", fontweight="heavy", fontsize=14)
plt.tight_layout()
plt.savefig("MVAnmdscar3.png", transparent=True, dpi=144)
plt.show()
```
